```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path("/Users/vaishak/Desktop/Predictive Maintenance for Aircraft Engines/airbus-predictive-maintenance/data/raw")  # adjust if your notebook path differs

# --- Column names for train/test ---
cols = (
    ['unit_number', 'time_in_cycles', 'op_setting_1', 'op_setting_2', 'op_setting_3']
    + [f'sensor_{i:02d}' for i in range(1, 22)]
)

# Use whitespace regex so multiple spaces are handled correctly
train_df = pd.read_csv(
    DATA_DIR / "train_FD001.txt",
    sep=r"\s+",
    header=None,
    names=cols,
    engine="python"
)

test_df = pd.read_csv(
    DATA_DIR / "test_FD001.txt",
    sep=r"\s+",
    header=None,
    names=cols,
    engine="python"
)

# RUL file: read first non-empty column only and name it 'RUL'
rul_df = pd.read_csv(
    DATA_DIR / "RUL_FD001.txt",
    sep=r"\s+",
    header=None,
    usecols=[0],          # take the first real column
    names=["RUL"],
    engine="python"
)

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
print("RUL shape:", rul_df.shape)

display(train_df.head())
display(test_df.head())
display(rul_df.head())
```
